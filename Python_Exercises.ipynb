```python
#AGE CALCULATER


# Import the datetime class from the datetime module.
from datetime import datetime
# Starts a try block to handle runtime errors.
try:
    birth_date_str = input("Enter your birth date (mm/dd/yyyy): ") # Prompts the user to enter the birth date.
    birth_date = datetime.strptime(birth_date_str, "%m/%d/%Y") # Converts the input string into a datetime.

    today = datetime.today() #takes the current date and time.

    age = today.year - birth_date.year

    print(f"Your age is: {age} years") #Displays the calculated age
    print("Birth date (European format):", birth_date.strftime("%d/%m/%Y")) #Converts the birth date into European format.

# Handles all possible Errors.
except ValueError:
    print("Error: Please enter date in mm/dd/yyyy format only.")

except Exception as e:
    print("Unexpected error occurred:", e)
```

```python
# PRIME NUMBER GENERATOR

# Defines a function named is_prime.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
# Starts a try block to handle runtime errors.
try:
    # Takes the start value and end value of the range, converts it to an integer.
    start = int(input("Enter start of range: "))
    end = int(input("Enter end of range: "))

    # Checks whether either input is zero or negative.
    if start <= 0 or end <= 0:
        raise ValueError("Inputs must be positive integers")

    # Creates an empty list to store the prime numbers.
    primes = []

    # Iterates through every number from start to end.
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)

    # Loops through the list in steps of 10 to display primes in rows of 10.
    print("Prime numbers:")
    for i in range(0, len(primes), 10):
        print(" ".join(map(str, primes[i:i+10])))

except ValueError as ve:
    print("Input Error:", ve) #Displays a meaningful error message for input-related issues.

except Exception as e:
    print("Unexpected error:", e) #Prints the error message to help identify the problem.
```

```python
#STUDENT MARKS PROCESSOR

#Opens a file named students.txt in write mode.
with open("students.txt", "w") as file:
    file.write("S023,87,76\n")
    file.write("S028,79,82\n")
    file.write("S032,83,89\n")
    file.write("S040,66,78\n")
    file.write("S059,59,67\n")
    file.write("S071,39,56\n")

#Creates an empty list to store processed student records.
students = []

#Starts a try block to handle errors.
try:
    with open("students.txt", "r") as f:
        for i, line in enumerate(f, 1):
            parts = line.strip().split(",")
            if len(parts) != 3:     #Checks whether the line has exactly three values.
                print(f"Skipping invalid line {i}: {line.strip()}") #Displays a message for invalid data format.
                continue
            reg, exam_str, course_str = parts #Assigns the values.

            #Starts a nested try block to handle numeric conversion errors.
            try:
                exam = float(exam_str)
                course = float(course_str)
            except ValueError:
                print(f"Skipping line {i} due to invalid numbers: {line.strip()}") #Displays an error for invalid numeric values.
                continue
            overall = exam * 0.7 + course * 0.3
            grade = "A" if overall >= 70 else "B" if overall >= 60 else "C" if overall >= 50 else "F" #Assigns a grade based on the overall marks.
            students.append((reg, exam, course, overall, grade))  #Stores the student record as a tuple.

    #Checks whether any valid student data was collected.
    if not students:
        print("No valid student data found.")
    else:
        students.sort(key=lambda x: x[3])

        with open("results.txt", "w") as f:
            for s in students:
                f.write(f"{s[0]} {s[3]:.2f} {s[4]}\n")

        grades_count = {}
        for s in students:
            grades_count[s[4]] = grades_count.get(s[4], 0) + 1

        print("Grade Statistics:")
        for grade in sorted(grades_count):
            print(f"{grade}: {grades_count[grade]}") #Prints each grade and the number of students who received it.

#Handles the case of errors.
except FileNotFoundError:
    print("File error: students.txt not found.")
except Exception as e:
    print("Unexpected error:", e)
```

```python
#WEATHER DATA FETCHER AND ANALYZER
import requests
import csv
from datetime import datetime

# Function to fetch weather data from OpenWeatherMap API
def fetch_weather(city: str, api_key: str) -> dict:
    try:
        url = "https://api.openweathermap.org/data/2.5/weather"
        params = {
            "q": city,
            "appid": api_key,
            "units": "metric"
        }

        response = requests.get(url, params=params)
        response.raise_for_status()  # Raises error for bad status codes
        return response.json()

    except requests.exceptions.HTTPError:
        print("Error: Invalid city name or API key.")
    except requests.exceptions.ConnectionError:
        print("Error: Network problem. Please check your internet.")
    except Exception as e:
        print("Unexpected error:", e)

    return None


# Function to analyze weather conditions
def analyze_weather(weather_data: dict) -> str:
    if not weather_data:
        return "No data to analyze"

    temp = weather_data["main"]["temp"]
    humidity = weather_data["main"]["humidity"]
    wind_speed = weather_data["wind"]["speed"]

    # Temperature category
    if temp <= 10:
        summary = "Cold (≤10°C)"
    elif 11 <= temp <= 24:
        summary = "Mild (11–24°C)"
    else:
        summary = "Hot (≥25°C)"

    # Warnings
    warnings = []
    if wind_speed > 10:
        warnings.append("High wind alert!")
    if humidity > 80:
        warnings.append("Humid conditions!")

    return summary + (" | " + " ".join(warnings) if warnings else "")


# Function to fetch weather and log it into a CSV file
def log_weather(city: str, filename: str):
    API_KEY = "9c99f079ed1aea6773fc505da02635a3"

    weather_data = fetch_weather(city, API_KEY)
    if not weather_data:
        return

    summary = analyze_weather(weather_data)

    temp = weather_data["main"]["temp"]
    humidity = weather_data["main"]["humidity"]
    wind_speed = weather_data["wind"]["speed"]
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    print("\nWeather Details")
    print("----------------")
    print("City:", city)
    print("Date & Time:", timestamp)
    print("Temperature:", temp, "°C")
    print("Humidity:", humidity, "%")
    print("Wind Speed:", wind_speed, "m/s")
    print("Summary:", summary)

    # Write data to CSV file
    with open(filename, "a", newline="") as file:
        writer = csv.writer(file)

        # Write header only if file is empty
        if file.tell() == 0:
            writer.writerow(["DateTime", "City", "Temperature", "Humidity", "WindSpeed", "Summary"])

        writer.writerow([timestamp, city, temp, humidity, wind_speed, summary])

    print("Weather data logged successfully!")


# Example usage
city_name = input("Enter city name: ")
log_weather(city_name, "weather_log.csv")
```
